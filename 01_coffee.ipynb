```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("coffee_cognition_realistic_700.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df['Brewing_Method'].value_counts()
```

```python
plt.figure()
sns.countplot(data=df,x='Brewing_Method')
plt.xticks(rotation=45)
plt.title('Brewing Method Distribution')
plt.show()
```

```python
plt.figure()
sns.countplot(data=df, x="Time_of_Day")
plt.title("Time of Day")
plt.show()
```

```python
sns.boxplot(x=df["Stroop_Reaction_Time_ms"])
plt.title("Outlier Inspection: Stroop Reaction Time")
plt.show()
```

```python
sns.boxplot(x=df["Caffeine_mg"])
plt.title("Outlier Inspection: Caffeine Intake")
plt.show()
```

```python
sns.scatterplot(
    x="Caffeine_mg",
    y="Stroop_Reaction_Time_ms",
    data=df
)
plt.title("Caffeine vs Stroop Reaction Time")
plt.show()
```

```python
sns.regplot(
    x="Caffeine_mg",
    y="Stroop_Reaction_Time_ms",
    data=df,
    scatter=False,
    color="red"
)

plt.title("Caffeine Intake vs Stroop Reaction Time")
plt.xlabel("Caffeine (mg)")
plt.ylabel("Stroop Reaction Time (ms)")
plt.show()
```

```python
sns.regplot(x="Caffeine_mg", y="Focus_Level", data=df, scatter=False, color='red')
plt.title("Caffeine mg vs Focus Level")
plt.xlabel("Caffeine mg")
plt.ylabel("Focus Level")
plt.show()
```

```python
sns.scatterplot(
    x="Sleep_Quality_Score",
    y="N_Back_Accuracy",
    data=df,
    alpha=0.6
)
sns.regplot(
    x="Sleep_Quality_Score",
    y="N_Back_Accuracy",
    data=df,
    scatter=False,
    color="red"
)
plt.title("Sleep Quality vs Working Memory Accuracy")
plt.show()
```

```python
numeric_cols = [
    "Age", "Caffeine_mg", "Sleep_Hours",
    "Sleep_Quality_Score", "Stress_Level",
    "Physical_Activity_Level",
    "Stroop_Reaction_Time_ms",
    "PVT_Reaction_Time_ms",
    "N_Back_Accuracy",
    "Focus_Level"
]

plt.figure(figsize=(10,8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Key Variables")
plt.show()
```
